```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.schema.runnable import RunnablePassthrough #두개의 출력을 쓰도록
chat = ChatOpenAI(
    model="gpt-3.5-turbo",
    temperature = 0.8
)
```

```python
poet_prompt = ChatPromptTemplate.from_messages([
    ("system","You're a fantastic poet. You're going to take a programming language and write a poem about it."),
    ("human","I want you to write a poem about {language}"),
])

poet_chain = poet_prompt | chat
poet_chain.invoke({
    "language": "python"
})
```

```python
poet_critic_prompt = ChatPromptTemplate.from_messages([
    ("system", "Display the entered poem. {poem} Write your review on the next line. You are a very professional critic who takes in poems and comments on them. But your reviews are not too long."),
    ("human", "{poem}")
])

critic_chain = poet_critic_prompt | chat
final_chain  = {"poem":poet_chain} | RunnablePassthrough.assign(review=critic_chain)
# 기본 poet_chain에서 poem을 생성하고 이것도 넘기고 또 이 입력을 바탕으로 review도 생성해라.
final_chain.invoke({
    "language": "python"
})

```
